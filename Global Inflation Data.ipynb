```python
df.duplicated().sum()
df = df.drop_duplicates()
print (df.head())
```

```python
df = df.dropna()
```

```python
df.fillna(method= 'ffill', inplace = True)
```

```python
df_long = df.melt(
    id_vars=["country_name", "indicator_name"],
    var_name="Year",
    value_name="Inflation Rate"
)
```

```python
df_long.head()
```

```python
df_long["Year"] = pd.to_numeric(df_long["Year"], errors='coerce')
```

```python
df_long = df_long.dropna(subset=["Inflation Rate"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_long["Inflation Rate"], bins=40, kde=True)
plt.title("Overall Global Inflation Distribution (1984â€“2024)")
plt.xlabel("Inflation Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
df_long["Inflation Rate"].max()
```

```python
df_long.sort_values("Inflation Rate", ascending=False).head()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df_long["Inflation Rate"])
plt.title("Inflation Rate Outlier Detection")
plt.show()
```

```python
recent_data = df_long[df_long["Year"] >= 2000]

sns.histplot(recent_data["Inflation Rate"], bins=30, kde=True)
plt.title("Inflation Distribution (2000â€“2024)")
plt.show()
```
